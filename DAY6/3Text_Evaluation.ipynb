```python
# BLUE SCORE
```

```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

reference = [["the", "capital", "of", "france", "is", "paris"]]
candidate = ["paris", "is", "the", "capital", "of", "france"]

score = sentence_bleu(reference, candidate, smoothing_function=SmoothingFunction().method1)
print("BLEU Score:", score)
```

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

scores = scorer.score("paris is the capital of france", "the capital of france is paris")
print(scores)
```

```python
# METEOR
```

```python
from nltk.translate.meteor_score import meteor_score
from nltk.tokenize import word_tokenize

reference = ["The capital of France is Paris"]
candidate = "Paris is the capital city of France"

tok_ref = [word_tokenize(ref) for ref in reference]
tok_cand = candidate.split()


score = meteor_score([reference], candidate)
print("METEOR Score:", score)
```

```python
# perplexity
```

```python
from transformers import GPT2LMHeadModel, GPT2TokenizerFast
import torch
import math

model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2TokenizerFast.from_pretrained("gpt2")

input_text = "Paris is the capital of France"
input_ids = tokenizer(input_text, return_tensors='pt')['input_ids']

with torch.no_grad():
    outputs = model(input_ids, labels=input_ids)
    loss = outputs.loss

perplexity = math.exp(loss)
print("Perplexity:", perplexity)
```
